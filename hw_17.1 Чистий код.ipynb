````python
def calculate_variance(numbers):
    """Обчислює дисперсію списку чисел."""
    if not numbers:
        return 0.0

    n = len(numbers)

    # Обчислюємо середнє значення
    mean = sum(numbers) / n

    # Обчислюємо середнє квадратів
    mean_squares = sum(x**2 for x in numbers) / n

    # Дисперсія = E[X^2] - (E[X])^2
    return mean_squares - (mean**2)
````
